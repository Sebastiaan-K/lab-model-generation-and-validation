```python
#If you find a pair of columns that show a high correlation between them (greater than 0.9), 
#drop the one that is less correlated with the column total_claim_amount. Write code for both the correlation matrix. 
#If there is no pair of features that have a high correlation, then do not drop any features.
```

```python
correlation_matrix = numerical.corr()

highly_correlated_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.62:
            highly_correlated_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j]))

columns_to_drop = set()
for col1, col2 in highly_correlated_pairs:
    corr_with_target_col1 = abs(correlation_matrix.loc[col1, 'Total Claim Amount'])
    corr_with_target_col2 = abs(correlation_matrix.loc[col2, 'Total Claim Amount'])
    if corr_with_target_col1 < corr_with_target_col2:
        columns_to_drop.add(col1)
    else:
        columns_to_drop.add(col2)

numerical = numerical.drop(columns=columns_to_drop)

correlation_matrix_updated = numerical.corr()
display("Updated Correlation Matrix:")
display(correlation_matrix_updated)
```

```python
#-- Numerical columns
#Did you see skewness in your numerical columns? -> If yess apply a Transfomer, and then an scaler
#-- Categorical columns
#Nominal -> OneHotEncoder
#Ordinal -> .replace() to replace the labels by numbers: 1)respecting the implicit order, 2)relative distance
#S -> 1 , M->2, L->3,...
```

```python
X_train_num = X_train_df.select_dtypes(np.number)
X_test_num  = X_test_df.select_dtypes(np.number)

X_train_cat = X_train_df.select_dtypes('object')
X_test_cat  = X_test_df.select_dtypes('object')

X_train_cat_nominal = X_train_cat[[]]
X_test_cat_nominal  = X_test_cat[[]]

X_train_cat_ordinal = X_train_cat[[]]
X_test_cat_ordinal  = X_test_cat[[]]
```

```python
#-- PowerTransformer(y)
#-- PowerTransformer(X)

from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler only on the training data
scaler.fit(X_train_num)

# Transform both training and test data
X_train_num_scaled = scaler.transform(X_train_num)
X_test_num_scaled = scaler.transform(X_test_num)
```

```python
X_train_transformed = np.concatenate([X_train_num_scaled, X_train_cat_nominal_encoded, X_train_cat_ordinal_encoded], axis=1)
X_test_transformed = np.concatenate([X_test_num_scaled, X_test_cat_nominal_encoded, X_test_cat_ordinal_encoded], axis=1)
```
